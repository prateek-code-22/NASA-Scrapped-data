## NeoWs (Near Earth Object Web Service) is a RESTful web service for near earth Asteroid information

### USING NASA NEOWS API TO EXTRACT LINKS , NUMBER OF ELEMENTS PASSED ,NEAR EARTH OBJECT LIKE INFORMATIONS.

```python
#importing libraries
import requests
from datetime import date, timedelta
import os
import urllib.request


#this api gives data of last 7 days so difference between starting and ending date must be of 7. 
start_date = date(2020, 7, 21)
end_date = date(2020, 7, 28)  

# store NASA API KEY to 'key' 
key = 'r8KSGXsGbEPvIRkAhYhyZqHvl01WjpQdUeXd1eqP'

#pass the parameter required for api.
params = {'start_date': start_date , 'end_date': end_date , 'api_key': key }

#get the request from nasa api server.
r = requests.get('https://api.nasa.gov/neo/rest/v1/feed?', params= params)    

#store data in .json file.
r = r.json()

#print the starting date and ending date.
print("STARTING DATE:",start_date)
print("ENDING DATE:  ",end_date)

#print new blank line.
print()

print("DETAIL LINKS INFORMATION:")
print("'next' for next 7 days")
print("'prev' for previous 7 days")
print("'self' for current 7 days")
print()

#printing the links
print("",r['links'])
print()

#number of element passed in 7 days.
print("NUMBER OF ELEMENT PASSED->",r['element_count'])
print()

#printing information about near earth objects. 
print("",r['near_earth_objects'])

#printing 80 stars.
print('*'*80)


```
